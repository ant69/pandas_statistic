# Домашнее задание к лекции «Случайные события, случайные величины»

Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

777: 200 руб.

999: 100 руб.

555: 50 руб.

333: 15 руб.

111: 10 руб.

*77: 5 руб.

**7: 3 руб.

*00: 2 руб.

**0: 1 руб.

### Вопросы:

Выгодна ли игра игроку?
Сколько в среднем приобретает или теряет игрок за одну игру?

### **Дополнительное задание повышенной сложности. **

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

– любая цифра

___
## Решение задания. Обязательная часть

```python
import numpy as np

# Фактически надо сравнить размер потраченных и выигранных средств
trials = paid = 1000000    # Количество попыток = потрачено рублей
earned = 0    # Выиграно рублей

for i in range(trials):
    rnd_trial = np.random.randint(0, 1000)    # Автомат выдает 
    if rnd_trial == 777:
        earned += 200
    elif rnd_trial == 999:
        earned += 100
    elif rnd_trial == 555:
        earned += 50
    elif rnd_trial == 333:
        earned += 15
    elif rnd_trial == 111:
        earned += 10
    elif rnd_trial % 100 == 77:
        earned += 5
    elif rnd_trial % 10 == 7:
        earned += 3
    elif rnd_trial % 100 == 0:
        earned += 2
    elif rnd_trial % 10 == 0:
        earned += 1
        
p_monte_carlo = earned / paid    # Вероятность из моделирования
# p = 2 / 6                             # Теоретическая вероятность

print(earned, paid, p_monte_carlo)
```

```python
# Теоретический расчет вероятности выигрыша (проигрыша)
# Вероятность выигрыша определяется как сумма произведений исходов (выигрышей) и вероятностей этих исходов

p_777 = 200 * (1 / 1000)
p_999 = 100 * (1 / 1000)
p_555 = 50 * (1 / 1000)
p_333 = 15 * (1 / 1000)
p_111 = 10 * (1 / 1000)
p_77 = 5 * (9 / 1000)         # 077 177 277 377 477 577 677    877 977
p_7 =  3 * (90 / 1000)        # 007 017 .. 067   087 097 107 .. 167   187 197 ..... 997
p_00 = 2 * (9 / 1000)
p_0 = 1 * (90 / 1000)

p_final = p_777 + p_999 + p_555 + p_333 + p_111 + p_77 + p_7 + p_00 + p_0

p_final
```
